### Colouring Geometrical Objects

So let's start with Points then LineSegments, Triangles and then Triples. As before we must first import the classes we need. There are no new classes this time, just some minor modifications which will be explained as we go along.

```python
from geom2d import Diagram
from geom2d import Point
from geom2d import LineSegment
from geom2d import Triangle
from phytrip import Triple 
```

### Colouring Points ###

Let's see how to colour a point. Note that in our code we use the word color, so as to be consistent with matplotlib, but in writing I prefer to stick with colour.

Any let us draw a point as before.

```python
diagram = Diagram()
p = Point(1,1)
diagram.add(p)
diagram.draw()
```

This just draws a blue point as before, since blue is the default colour for points. If we want to draw a green point in a different place we do so as follows.

```python
p_green = Point(2, -2, color='green')
diagram.add(p_green)
diagram.draw()
```

Similarly for a red point at (4, 4) we do the following.

```python
p_red = Point(4, 4, color='red')
diagram.add(p_red)
diagram.draw()
```

### Colouring Line Segments ### 

Ok. We now do the same thing for LineSegments. Firstly clear the diagram, and then create a linesegment object with the default colour, then we add it to the diagram and draw it.

```python
diagram = Diagram()
l = LineSegment(Point(0,0), Point(2,2))
diagram.add(l)
diagram.draw()
```

So as maybe expected we get a blue line with blue end-points. The next volution of this software will allow us to have different coloured points and lines, but for the moment we will stick with this simple approach.

Next we are going to draw read and green lines in our diagram.

```python
l_red = LineSegment(Point(-2,0), Point(0,2), color='red')
diagram.add(l_red)

l_green = LineSegment(Point(2,0), Point(4,2), color='green')
diagram.add(l_green)
diagram.draw()
```

I have just realised that I need arrows and text. I will leave that for the moment.

###Colouring Triangles###

Again the scheme is consistent. The default color is blue and red and or green colours can be used. So let's explore a little.

```python
diagram = Diagram()
t = Triangle(Point(0,0), Point(3,0), Point(3,4))
diagram.add(t)
diagram.draw()
```

```python
t_red = Triangle(Point(0,0), Point(-3,0), Point(-3,4), color='red')
diagram.add(t_red)
diagram.draw()
```

```python
t_green = Triangle(Point(0,0), Point(2, -2), Point(-2,-4), color='green')
diagram.add(t_green)
diagram.draw()
```

So we see that it is very easy to get a consistent colouring scheme for geometrical objects, and we are not limitd to the colours that we have used, as evidenced by the range of named colors available in matplotlib. They are discussed in the [list of named colors](https://matplotlib.org/stable/gallery/color/named_colors.html). Personally I feel that, at this stage three colours is more than enough.

### Colouring Triples ###

So what about triples. Obviously they are just a flavour of triangles, so are fully colourable. We declare the desired color in the make_triangle method as follows.

```python
diagram = Diagram()
t = Triple(4, 3, 5)
tri = t.make_triangle(color='red')
diagram.add(tri)
diagram.draw()
```

The default colour is blue as before, so we have over-ridden it with red.

So that's it for the moment.
