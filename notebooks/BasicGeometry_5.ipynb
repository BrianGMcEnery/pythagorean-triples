## Drawing Rectangles and Circles ##

This time we have two new classes to import, one for rectangles and the other for circles.

```python
from geom2d import Diagram
from geom2d import Point
from geom2d import LineSegment
from geom2d import Triangle
from geom2d import Rectangle
from geom2d import Circle
from phytrip import Triple 
```

### Drawing Rectangles ###

Rectangles are specified by two Points, the first specifies the bottom left corner, and the second, the top right corner. They are created as follows.

```python
diagram = Diagram()
r = Rectangle(Point(0, 0), Point(1, 1))
diagram.add(r)
diagram.draw()
```

Obviously we can create red rectangles as well, and green. We can also rotate them, and translate them, just as with any other geometrical object.

```python
r_red = Rectangle(Point(2,0), Point(4, 2), color='red')
diagram.add(r_red)
r_green = Rectangle(Point(-4,-4), Point(-2, -2), color='green')
diagram.add(r_green)
diagram.draw()
```

Let's clear the diagram, then rotate the green rectangle by $45^o$ about the origin, and see what happens when we draw the diagram.

```python
diagram = Diagram()
r_green_45 = r_green.rotate(45)
diagram.add(r_green_45)
diagram.draw()
```

Oops! The rotated rectangle extends beyond the limits of our drawing. This is easily resolved by changing the default arguments to the class creation function Diagram as follows.

```python
diagram = Diagram(xlim=(-6, 6), ylim=(-6, 6))
```

The limits of the new drawing are extended and the rotate rectangle will fit, as required.

```python
diagram.add(r_green_45)
diagram.draw()
```

So much for rectangles, what about circles.

### Drawing Circles ###

Circles are specified by the origin as a Point object and the radius.

```python
diagram = Diagram()
c = Circle(Point(0,0), 1)
print(c)
diagram.add(c)
diagram.draw()
```

A unit circle at the origin.

We can translate, rotate and colour as before.

```python
c_new = c.translate(Point(2,0))
c_new = c_new.rotate(45)
c_new.set_color('red')
diagram.add(c_new)
diagram.draw()
```

This series of steps does the following
* translate the origin of the circle to (2,0)
* rotate the circle by $45^o$ about the origin
* set the color of the new circle to red

The only new thing here is using the instance method set_color('red'). This will ensure that the points associated with the geometrical object are coloured in the same way as the object itself.
